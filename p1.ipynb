```python
class node:
    def __init__(self,val):
        self.val=val
        self.left=None
        self.right=None

drinks=node("drinks")
hot=node("hot")
cold=node("cold")
tea=node("tea")
coffee=node("coffee")
cola=node("cola")
fanta=node("fanta")

hot.left=tea
hot.right=coffee
cold.left=cola
cold.right=fanta
drinks.left=hot
drinks.right=cold

print(drinks.left.val)
```
