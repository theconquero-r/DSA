```python
# queue

class queue:
    def __init__(self):
        self.items=[]
    
    def is_empty(self):
        return len(self.items)==0
    def enqueue(self,item):
        self.items.append(item)
    
    def dequeue(self):
        if len(self.items)==0:
            print("queue is empty")
            return
        x=self.items.pop(0)
        return x
    def front(self):
        if len(self.items)==0:
            print('queue is empty')
            return
        return self.items[0]
    def rear(self):
        if len(self.items)==0:
            print('queue is empty')
            return
        return self.items[-1]
    def size(self):
        return len(self.items)
    
q=queue()
q.enqueue(23)
q.enqueue(5)
q.enqueue(15)
q.enqueue(30)
q.size()
q.front()
q.rear()
q.dequeue()
q.dequeue()
q.front()
# q.rear()
# q.is_empty()
```

```python
from collections import deque

class node:
    def __init__(self,val):
        self.val=val
        self.left=None
        self.right=None

five= node(5)
three= node(3)
two= node(2)
nine= node(9)
four= node(4)
eight= node(8)
ten= node(10)
one= node(1)
six= node(6)
five.left=three
three.left=two
three.right=nine
five.right=four
four.left=eight
four.right=ten
eight.left=one
eight.right=six

def level_order(root):
    result=[]
    queue=deque([])
    queue.append(root)
    while (len(queue))!=0:
        e=queue.popleft()
        result.append(e.val)
        if e.left!=None:
            queue.append(e.left)
        if e.right!=None:
            queue.append(e.right)
    return result

level_order(five)
```

```python
from collections import deque
values = [5,3,4,2,9,8,10,1,6]

class node:
    def __init__(self,val):
        self.val=val
        self.left=None
        self.right=None

def build_tree(values):
    if not values:
        return None

    root = node(values[0])
    queue = deque([root])
    i = 1

    while i < len(values):
        current = queue.popleft()

        # left child
        if i < len(values):
            current.left = node(values[i])
            queue.append(current.left)
            i += 1

        # right child
        if i < len(values):
            current.right = node(values[i])
            queue.append(current.right)
            i += 1

    return root


root = build_tree(values)
```
