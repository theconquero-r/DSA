```python
#deleting a node:
#insertion 
values=[9,6,13,3,8,11,14,2,4,7,9,None,None,None,None,None,None,None,5]

from collections import deque
class node:
    def __init__(self,root):
        self.val=root
        self.left=None
        self.right=None

def build_tree(values):
    if not values:
        return
    root=node(values[0])
    queue=deque([root])
    i=1
    while queue and i<len(values):
        curr=queue.popleft()
        if i<len(values) and values[i] is not None:
            curr.left=node(values[i])
            queue.append(curr.left)
        i+=1

        if i<len(values) and values[i] is not None:
            curr.right=node(values[i])
            queue.append(curr.right)
        i+=1

    return root

root=build_tree(values)


def insert(root,key):
    if not root:
        return node(key)
    
    temp=root
    while True:
        if key<temp.val:
            if temp.left is None:
                temp.left=node(key)
                break
            else:
                temp=temp.left
        else:
            if temp.right is None:
                temp.right=node(key)
                break
            else:
                temp=temp.right
    return root

# root=insert(root,12)

#count number of nodes
def deletion(node):
    if node.left==None:
        return node.right
    if node.right==None:
        return node.left
    else:
        right_child=node.right
        last_right=findlastright(node.left)
        last_right.right=right_child
        return node.left
def findlastright(node):
    while node.right!=None:
        node= node.right
    return node
def delete_node(root,key):
    if not root:
        return None
    if root.val==key:
        return deletion(root)
    temp=root
    while temp !=None:
        if temp.val>key:
            if temp.left is not None and temp.left.val==key:
                
                temp.left=deletion(temp.left)
            else:
                temp=temp.left
        else:
            if temp.right is not None and temp.right==key:
                temp.right=deletion(temp.right)
            else:
                temp=temp.right
        return root

delete_node(root,6)

def level_order(root):
    if root==None:
        return 0
    queue=deque([root])
    result=[]
    while queue:
        curr=queue.popleft()
        result.append(curr.val)
        if curr.left:
            queue.append(curr.left)
        if curr.right:
            queue.append(curr.right)
    return result

level_order(root)

# time complexity is o(log2n)
# space complexity is O(1)
```
