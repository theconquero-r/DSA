#BINARY SEARCH TREE

SEARCHING IN BINARY TREE

```python
from collections import deque
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
values=[9,3,11,2,7,None,15,None,None,4,8,14,None]

def build_tree(values):
    if not values:
        return None
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    while queue and i < len(values):
        node = queue.popleft()
        if values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    return root

root = build_tree(values)

def findTarget(root, k):
    temp=root
    while temp!=None:
        if temp.val==k:
            return True
        elif temp.val>k:
            temp=temp.left
        else:
            temp=temp.right
    return False
findTarget(root, 10)
# time complexity: O(log2 n)
# space complexity: O(1)
```

```python
#find minimum value in a binary search tree
def findMin(root):
    temp=root
    while temp.left!=None:
        temp=temp.left
    return temp.val
findMin(root)

# time complexity: O(log2 n)
# space complexity: O(1)
```

```python
# find the ceil value in a binary search tree
def findCeil(root, k):
    ceil=-1
    while root!=None:
        if root.val==k:
            return root.val
        elif root.val<k:
            root=root.right
        else:
            ceil=root.val
            root=root.left
    return ceil
findCeil(root, 6)

# time complexity: O(log2 n)
# space complexity: O(1)
```

```python
# find floor value in a binary search tree
def findFloor(root, k):
    floor=-1
    while root!=None:
        if root.val==k:
            return root.val
        elif root.val<k:
            floor=root.val
            root=root.right
        else:
            root=root.left
    return floor
findFloor(root, 6)
# time complexity: O(log2 n)
# space complexity: O(1)
```
