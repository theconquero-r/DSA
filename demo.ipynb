```python
values=[9,3,11,2,7,None,15,None,None,4,8,14,None]
```

```python
from collections import deque
class node:
    def __init__(self,root):
        self.val=root
        self.left=None
        self.right=None

def build_tree(values):
    if not values:
        return
    root=node(values[0])
    queue=deque([root])
    i=1
    while queue and i<len(values):
        curr=queue.popleft()
        if i<len(values) and values[i] is not None:
            curr.left=node(values[i])
            queue.append(curr.left)
        i+=1

        if i<len(values) and values[i] is not None:
            curr.right=node(values[i])
            queue.append(curr.right)
        i+=1

    return root

root=build_tree(values)
```

```python
#count number of nodes

def count(root):
    if root==None:
        return 0
    queue=deque([root])
    nodes=0
    while queue:
        curr=queue.popleft()
        nodes+=1
        if curr.left:
            queue.append(curr.left)
        if curr.right:
            queue.append(curr.right)
    return nodes

count(root)
```

```python
#height of a binary tree:

def height_rec(root):
    if not root:
        return -1
    
    left_height=height_rec(root.left)
    right_height=height_rec(root.right)
    return max(left_height,right_height)+1

print(height_rec(root))


def height_iterative(root):
    h=-1
    
    queue=deque([root])
    while queue:
        len_queue=len(queue)
        h+=1
        for i in range(len_queue):
            curr=queue.popleft()
            if curr.left!=None:
                queue.append(curr.left)

            if curr.right!=None:
                queue.append(curr.right)
            
    return h

height_iterative(root)
```

```python
#check if a tree is balanced or not

def check_balance(root):
    if not root:
        return -1
    left_depth=check_balance(root.left)
    if left_depth==-2:
        return -2
    right_depth=check_balance(root.right)
    if right_depth==-2:
        return -2
    if abs(left_depth-right_depth)>1:
        return -2
    return max(left_depth,right_depth)+1

if check_balance(root)!=-2:
    print("tree is balanced")
else:
    print("tree is not balanced")
```

```python
#search in binary tree

def search(root,key):
    if not root:
        return None
    
    while root!=None:
        if root.val==key:
            return root.val
        elif root.val<key:
            root=root.right
        else:
            root=root.left

search(root,7)
```
