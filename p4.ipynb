```python
from collections import deque
class node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None

values=[3,3,1,1,2,3,6,None,None,None,None,1,None,None,9,1,2,None,10,6,None,None,None,5,None,None]

def build_tree(values):
    if not values:
        return
    root=node(values[0])
    queue=deque([root])    
    i=1
    while queue and i<len(values):
        curr=queue.popleft()
        if i<len(values) and values[i] is not None:
            curr.left=node(values[i])
            queue.append(curr.left)
        i+=1
        if i<len(values) and values[i] is not None:
            curr.right=node(values[i])
            queue.append(curr.right)
        i+=1
    return root

root=build_tree(values)

```

```python
#traversal of binary tree
def level_order(root):
    if not root:
        return
    queue=deque([])
    queue.append(root)
    result=[]
    while queue:
        curr=queue.popleft()
        result.append(curr.data)
        if curr.left!=None:
            queue.append(curr.left)

        if curr.right!=None:
            queue.append(curr.right)
    return result

level_order(root)
```

```python
from collections import deque
class node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None

values=[3,3,1,1,2,3,6,None,None,None,None,1,None,None,9,1,2,None,10,6,None,None,None,5,None,None,8]

def build_tree(values):
    if not values:
        return
    root=node(values[0])
    queue=deque([root])    
    i=1
    while queue and i<len(values):
        curr=queue.popleft()
        if i<len(values) and values[i] is not None:
            curr.left=node(values[i])
            queue.append(curr.left)
        i+=1
        if i<len(values) and values[i] is not None:
            curr.right=node(values[i])
            queue.append(curr.right)
        i+=1
    return root

root=build_tree(values)


def diameter(root):
    diameter=0
    def height(node):
        nonlocal diameter
        if node==None:
            return -1
        left_height=height(node.left)
        right_height=height(node.right)
        diameter=max(diameter,left_height+right_height+2)
        return max(left_height,right_height)+1
    height(root)
    return diameter
diameter(root)
```

```python
from collections import deque

def print_levels(root):
    q = deque([root])   # ek queue banayi aur root node daal diya
    while q:            # jab tak queue khali na ho
        n = len(q)      # current level ke nodes kitne hain
        level_has_node = False
        level_nodes = []
        for _ in range(n):   # har node ko process karo
            x = q.popleft()  # queue se ek node nikalo
            if x:            # agar node exist karta hai
                level_nodes.append(str(x.data))   # uska data print list mein daalo
                q.append(x.left)                  # uske left child ko queue mein daalo
                q.append(x.right)                 # uske right child ko queue mein daalo
                level_has_node = True             # matlab is level mein real node hai
            else:
                level_nodes.append("x")           # agar node None hai to "x" daalo
        # agar poore level mein ek bhi real node nahi mila, stop kar do
        if not level_has_node:
            break
        print(" ".join(level_nodes))   # current level print karo

print_levels(root)
```
